ðŸ“Š **JustificaciÃ³n del Outlier 'duration'**

```python
# justificacion outlier duration
# Calcular el Cuartil 3 (Q3) y el Rango IntercuartÃ­lico (IQR) para la duraciÃ³n
Q1_dur = df['duration(imp:median)'].quantile(0.25)
Q3_dur = df['duration(imp:median)'].quantile(0.75)
IQR_dur = Q3_dur - Q1_dur

# Definir el lÃ­mite superior (ejemplo de acotamiento: 1.5 * IQR)
limite_superior_outlier = Q3_dur + 1.5 * IQR_dur

# Identificar el conteo de Outliers por encima del lÃ­mite
conteo_outliers_duracion = len(df[df['duration(imp:median)'] > limite_superior_outlier])

print(f"LÃ­mite superior calculado (1.5 * IQR): {limite_superior_outlier:.2f} segundos")
print(f"NÃºmero de Outliers de duraciÃ³n por encima de este lÃ­mite: {conteo_outliers_duracion}")
print(f"El valor mÃ¡ximo de duraciÃ³n es: {df['duration(imp:median)'].max()} segundos (casi 82 minutos)")
```

ðŸ“Š **JustificaciÃ³n del Outlier 'campaign'**

```python
# justificacion out campaign
# Calcular el Cuartil 3 (Q3) y el Rango IntercuartÃ­lico (IQR) para campaign
Q1_camp = df['campaign'].quantile(0.25)
Q3_camp = df['campaign'].quantile(0.75)
IQR_camp = Q3_camp - Q1_camp

# Definir el lÃ­mite superior (ejemplo de acotamiento: 1.5 * IQR)
limite_superior_outlier_camp = Q3_camp + 1.5 * IQR_camp

# Identificar el conteo de Outliers por encima del lÃ­mite
conteo_outliers_campaign = len(df[df['campaign'] > limite_superior_outlier_camp])

print(f"LÃ­mite superior calculado (1.5 * IQR) para campaign: {limite_superior_outlier_camp:.2f} contactos")
print(f"NÃºmero de Outliers de campaÃ±a por encima de este lÃ­mite: {conteo_outliers_campaign}")
print(f"El valor mÃ¡ximo de campaign es: {df['campaign'].max()} contactos")
```

```python
df['efficiency_index'] = df['success_flag'] / (df['duration'] + 1)
```

```python
import pandas as pd
import numpy as np

# --- 1ï¸âƒ£ Variable: Grupo de edad ---------------------------------
# SegmentaciÃ³n basada en madurez financiera
df['age_group'] = pd.cut(
    df['age'],
    bins=[0, 30, 50, 100],
    labels=['Joven', 'Adulto', 'Mayor']
)

# --- 3ï¸âƒ£ Variable: Trimestre de contacto --------------------------
# Mapeo de mes â†’ trimestre
month_to_quarter = {
    'jan': 'Q1', 'feb': 'Q1', 'mar': 'Q1',
    'apr': 'Q2', 'may': 'Q2', 'jun': 'Q2',
    'jul': 'Q3', 'aug': 'Q3', 'sep': 'Q3',
    'oct': 'Q4', 'nov': 'Q4', 'dec': 'Q4'
}
df['quarter'] = df['month'].map(month_to_quarter)

# --- 4ï¸âƒ£ Variable: Bandera de Ã©xito (0/1) --------------------------
df['success_flag'] = df['y'].map({'yes': 1, 'no': 0})

# --- 5ï¸âƒ£ Variable: Ãndice de eficiencia ----------------------------
# Mide el Ã©xito en funciÃ³n del tiempo invertido
df['efficiency_index'] = df['success_flag'] / (df['duration'] + 1)

# --- 6ï¸âƒ£ Variable: Intensidad de contacto --------------------------
# Suma de contactos en campaÃ±a actual + campaÃ±as previas
df['contact_intensity'] = df['campaign'] + df['previous']

# --- 7ï¸âƒ£ Variable: Tasa de Ã©xito previa del cliente ----------------
# CuÃ¡ntas veces aceptÃ³ / nÃºmero de contactos anteriores
df['past_success_rate'] = df['previous'] / (df['campaign'] + 1)

# Evitar divisiÃ³n por cero o NaN
df['past_success_rate'] = df['past_success_rate'].fillna(0)

# --- 8ï¸âƒ£ Variable: Historial de contacto reciente ------------------
df['recent_contact'] = df['pdays'].apply(
    lambda x: 'No contactado antes' if pd.isna(x) else 'Contactado antes'
)

# ---------------------------------------------------------------
# Vista previa de las nuevas variables creadas
df[['age_group','quarter','efficiency_index',
    'contact_intensity','past_success_rate','recent_contact']].head()
```
